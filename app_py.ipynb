```python
import gradio as gr
```

```python
from transformers import pipeline
```

```python
#load the pre trained sentiment analysis model
sentiment_analyzer = pipeline("sentiment-analysis")
```

```python
#define the funtion that will wrao our model
def analyze_sentiment(text):
  result = sentiment_analyzer(text)[0]
  #gradio works best with dictionary outputs
  return {result['label']: result ['score']}
```

```python
#create the gradio interface
iface= gr.Interface(
    fn=analyze_sentiment,
    inputs = gr.Textbox(lines=2, placeholder="Enter a sentence here.."),
    outputs="label",
    title="Sentiment Analysis Bot",
    description =' Type in a sentence and see if the model thinks its positive or negative.Built with Gradio and Hugging Face'
)
#launch the app
iface.launch(share=True)
```
